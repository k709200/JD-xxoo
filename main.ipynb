```python
from bs4 import BeautifulSoup 
from selenium import webdriver
import requests as req

jd_dir="/home/k709200/Downloads/jd/"
index = ('http://jandan.net/ooxx/MjAyMTAzMDgtMTE5')
 

def main(urls):
    driver=webdriver.Chrome('/home/k709200/Documents/chromedriver')
    driver.maximize_window()
    #print ("{}".format(url))
    driver.get(urls)
    data = driver.page_source
    # print(data)
    soup = BeautifulSoup(data,'html.parser')
    # print(soup.title.string)
    grades = soup.find_all('a')
    driver.close()

    try:
        for i in grades:
            t=i.get('href')
            if 'large' in t:
                pic_name=t.split('/large/')[1]
                print(t)
                # write pic
                pic=req.get("http:"+t)
                pic_file=open(jd_dir+pic_name, 'wb')
                pic_file.write(pic.content)
                pic_file.close()
    except:
        print('done')

    
    try:
        next_no=1
        for j in grades:
            t2=j.get('href')
            if '//jandan.net/ooxx/'in t2:
                next_no=next_no+1
                if next_no==5:
                    return ('http:'+t2)
                    print('http:'+t2)
    except:
        print('url ok')
        
        
    


next_url=main(index)
for i in range(119):
    print(next_url)
    next_url=main(next_url)
    
```
